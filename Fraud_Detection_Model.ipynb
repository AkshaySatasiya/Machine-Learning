## Scaling

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

df_new['scaled_amount'] = std_scaler.fit_transform(df_new['Amount'].values.reshape(-1,1))
df_new['scaled_time'] = std_scaler.fit_transform(df_new['Time'].values.reshape(-1,1))

df_new.drop(['Time','Amount'], axis=1, inplace=True)
```

```python
df_new
```

```python
scaled_amount = df_new['scaled_amount']
scaled_time = df_new['scaled_time']

df_new.drop(['scaled_amount','scaled_time'], axis=1, inplace=True)
df_new.insert(0, 'scaled_amount', scaled_amount)
df_new.insert(1, 'scaled_time', scaled_time)

df_new.head()
```

hTe __"time"__ and __"amount"__ columns were observed to have different scales compared to other features. To solve this we have used the __"StandardScaler" technique__ to scaled down the values of these columns. 

## Handling Imbalanced Dataset [Using SMOTE]

```python
df_new.value_counts('Class')
```

As We can see inthe above cell that the dataset is imbalanced. the count of __0 = 284315__ and the count of __1 = 492__ only. 

Now we are going to use the __SMOTE__ technique to balance the dataset.

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)

X = df_new.drop('Class',axis=1)
y = df_new['Class']

X_res,y_res = sm.fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
len(y_res)
```

```python
len(X_res)
```

## Train Test Split

```python
X_res.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(
    X_res,y_res,test_size=0.2,random_state=42)
```

# Model Training

Here we are going to train multiple model. So that we can get to know that which model performs better and we can use that model for further use.

So we are going to train the following models : 

    1. Logistic Regression
    2. Naive Bayes
    3. K-Nearest Neighbour
    4. Decision Tree
    5. Random Forest

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

models = {
    'Logistic Regression':LogisticRegression(),
    'Naive Bayes Classifier':GaussianNB(),
    'KNN Classifier':KNeighborsClassifier(),
    'Decision Tree Classsifier':DecisionTreeClassifier(),
    'Random Fores Classifier ':RandomForestClassifier()
}


for i in range(len(list(models))):
    model = list(models.values())[i]
    print(list(models.keys())[i])
    print("\n")

    
    cv_scores = cross_val_score(model, X_train, y_train, cv=5)  
    print("Cross-validation scores:", cv_scores)
    print(f"Average CV accuracy for {list(models.keys())[i]}:", np.mean(cv_scores))


    # Training of the model
    model.fit(X_train, y_train)

    # predictions
    y_pred = model.predict(X_test)

    # Evaluation
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    
    print(f"Test set accuracy for {(list(models.keys())[i])}:", accuracy)
    print("Confusion Matrix:\n", conf_matrix)
    print("Classification Report:\n", class_report)
    print("\n" + "="*50 + "\n")
```

Here we evaluated the performance of five distinct algorithms: Logistic Regression, Naive Bayes, K Nearest Neighbors, Decision Tree, and Random Forest. The assessment included accuracy metrics and cross-validation scores.
